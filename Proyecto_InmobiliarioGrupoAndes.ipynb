# 📊 Proyecto: Análisis Comercial Inmobiliario

El sector inmobiliario necesita evaluar su **desempeño comercial** para comprender su crecimiento, rentabilidad y comportamiento de clientes.

Se construira un **dashboard ejecutivo** que permita analizar ventas, clientes y propiedades para apoyar decisiones estratégicas basadas en datos.


# 📂 Descripción de los Datasets

🔹 **hecho_ventas_propiedades**  
Tabla de hechos con las transacciones de venta (precio, cliente, propiedad, canal, fecha).

🔹 **dim_clientes**  
Información y segmentación de clientes.

🔹 **dim_propiedades**  
Características de las propiedades (tipo, tamaño, ubicación, etc).

🔹 **dim_fecha**  
Tabla calendario que deberá ser creada como parte del modelado.

# 🎯 Preguntas que el dashboard debe responder

**Desempeño general**

- ¿Cuál es el **ingreso total** generado por la empresa?
- ¿Cuántas **propiedades se han vendido**?
- ¿Cuál es el **precio promedio de venta**?
- ¿Cuál es la **comisión total** generada por la empresa?

**Análisis comercial**

- ¿Qué **tipo de propiedad genera más ingresos**?
- ¿Qué **segmento de clientes compra más propiedades**?
- ¿Qué canal de venta **genera mayor ingresos**?


**Análisis temporal**

- ¿Cómo evolucionan las **ventas en el tiempo**?
- ¿El negocio está **creciendo año contra año**?
- ¿Cómo va el desempeño **acumulado del año actual (YTD)**?

**Cohortes de clientes**

- ¿Los clientes **vuelven a comprar** después de su primera compra?
- ¿Qué **cohortes de clientes generan más compras/ingresos con el tiempo**?

## 📝Paso 6:  Resumen Ejecutivo

**Hallazgos clave**
- El ingreso total del periodo fue de $6,012,502,170 generado por 8,500 ventas entre enero 2023 y diciembre 2024. El negocio muestra una trayectoria de crecimiento sostenida, cerrando 2024 con +11.1% más ingresos y +10.1% más volumen de ventas respecto a 2023.


- El tipo de propiedad que genera mayor revenue es Casa con 2.241M (37.26%),aunque Comercial tiene el ticket promedio más alto del portafolio con $1,797,097 por transacción — casi 5 veces el ticket de Departamento.


- La ciudad con mayor volumen de ventas es Ciudad de México con 3.242M (53.9% del total), seguida de Bogotá con $2.770M (46.1%).


- El canal de venta más eficiente en términos de ingresos es Corredor, que concentra el 72.8% del revenue (4.380M) con 6,181 transacciones. La diferencia con el canal Directo no es de precio — ambos tienen ticket similar (~$707K) — sino de volumen.


**Métricas principales**

- Cantidad de Ventas= 8,500.
- Ticket Promedio= $707,353.


- Comisión Total = $200,6 M.
- Crecimiento YoY= +11.1.
- Clientes únicos= 3,139.
- Clientes con recompra = 2,421 (77.1%).
- Promedio compras/cliente= 2.71

**Insights accionables**


- El segmento "Primera vez" concentra el 62.9% del ingreso ($3.784M) con 5,302 ventas. Son el motor principal del negocio en adquisición, pero su retención mensual promedio es de solo 9.4% — lo que significa que la mayoría no vuelve a comprar en el corto plazo. Esta es la brecha más importante a cerrar.
- Las cohortes de Mar y Abr 2023 son las más grandes con 295,608,036 y 244,940,882 primeras compras respectivamente, y generan los mayores ingresos acumulados del periodo. Los clientes adquiridos en H1 2023 muestran mayor actividad histórica por tener más meses de vida, lo que confirma que la retención mejora con el tiempo si se mantiene el contacto.
- Las ventas tienen estacionalidad clara: los picos de ingresos se repiten consistentemente en Q1 y Q3 en ambos años. Esto es una señal de demanda estacional que puede aprovecharse para planificar campañas y recursos comerciales.
- El 77.1% de los clientes vuelve a comprar con un promedio de 2.71 transacciones por cliente — una base de recompra sólida que indica confianza en el producto. El reto es acortar el tiempo entre la primera y segunda compra.

**Recomendaciones estratégicas**

1. Programa de retención para compradores "Primera vez" — prioridad alta
Implementar una secuencia de contacto en los primeros 90 días post-cierre orientada a la segunda compra. Elevar la retención mensual de 9.4% a 15% podría representar aproximadamente $180M adicionales en ingresos anuales.


3. Priorizar Casa y Comercial en el portafolio — prioridad alta
Casa genera el mayor ingreso total y Comercial el mayor ticket por transacción. Concentrar los esfuerzos de prospección y cierre en estos dos tipos maximiza el retorno por venta.


5. Crecer el canal Directo en propiedades Comerciales — prioridad media
El canal Directo tiene ticket similar al Corredor pero solo representa el 27.2% del revenue. En el segmento Comercial, cada venta directa evita pagar comisión sobre $1.8M — el impacto en margen neto es significativo.



7. Planificar campañas en Q1 y Q3 — prioridad media
Los picos estacionales son consistentes y predecibles. Reforzar la fuerza comercial y los presupuestos de marketing en enero–marzo y julio–septiembre maximiza la conversión en los meses de mayor demanda natural.



9. Estrategia diferenciada por ciudad — prioridad baja
Ciudad de México y Bogotá tienen perfiles de demanda distintos. Desarrollar un mix de producto y canal específico para cada mercado puede optimizar el ingreso en ambas plazas sin aumentar el costo de operación.
